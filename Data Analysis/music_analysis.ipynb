```python
df['decade'] = (df['year'] // 10) * 10
```

```python
df.describe()
```

```python
df.head()
```

```python
energy_by_decade = df.groupby('decade')['Energy'].agg(['mean', 'median', 'std'])
print("\nEnergy by Decade:")
print(energy_by_decade)
```

```python
# Visualize
plt.figure(figsize=(12, 6))
plt.plot(energy_by_decade.index, energy_by_decade['mean'], marker='x', linewidth=2, markersize=8)
plt.title('Average Song Energy by Decade', fontsize=16, fontweight='bold')
plt.xlabel('Decade', fontsize=12)
plt.ylabel('Average Energy', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
dance_by_year = df.groupby('year')['dance'].mean()
```

```python
# Calculate overall trend
years = df['year'].values
dances = df['dance'].values
```

```python
# Remove NaN values for correlation
mask = ~np.isnan(dances)
correlation = np.corrcoef(years[mask], dances[mask])[0, 1]
```

```python
print(f"\nTempo Trend: Correlation between year and tempo = {correlation:.3f}")
if correlation > 0.1:
    print("Music is getting FASTER over time")
elif correlation < -0.1:
    print("Music is getting SLOWER over time")
else:
    print("Tempo has remained relatively STABLE over time")
```

```python
# Visualize
plt.figure(figsize=(14, 6))
plt.plot(dance_by_year.index, dance_by_year.values, alpha=0.6, linewidth=1)
# Add trend line
z = np.polyfit(dance_by_year.index, dance_by_year.values, 1)
p = np.poly1d(z)
plt.plot(dance_by_year.index, p(dance_by_year.index), "r--", linewidth=2, label='Trend Line')
plt.title('Average Dance of Songs Over Time', fontsize=16, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Dance (BPM)', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# ============================================================
# 2. GENRE COMPARISON
# ============================================================

print("\n" + "=" * 60)
print("2. COMPARING GENRES")
print("=" * 60)
```

```python
danceability_by_genre = df.groupby('top genre')['dance'].mean().sort_values(ascending=False)

print("\nTop 10 Most Danceable Genres:")
print(danceability_by_genre.head(10))
```

```python
# Question: How do acoustic songs compare to electronic ones?
# ------------------------------------------------------------

# Define acoustic vs electronic 
```

```python
df['song_type'] = 'Mixed'
df.loc[df['acous'] > 50, 'song_type'] = 'Acoustic'
df.loc[df['acous'] < 20, 'song_type'] = 'Electronic'
```

```python
# Compare multiple features
comparison_features = ['Energy', 'dance', 'val']
comparison = df.groupby('song_type')[comparison_features].mean()

print("\nAcoustic vs Electronic Comparison:")
print(comparison)
```

```python
# Visualize comparison
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for idx, feature in enumerate(comparison_features):
    ax = axes[idx]
    comparison[feature].plot(kind='bar', ax=ax, color=['#2ecc71', '#3498db', '#e74c3c'])
    ax.set_title(f'Average {feature.capitalize()}', fontsize=12, fontweight='bold')
    ax.set_ylabel(feature.capitalize())
    ax.set_xlabel('')
    ax.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

```python
# ============================================================
# 3. POPULARITY PATTERNS
# ============================================================

print("\n" + "=" * 60)
print("3. ANALYZING POPULARITY PATTERNS")
print("=" * 60)
```

```python
# Question 3a: What features do popular songs have in common?
# ------------------------------------------------------------

# Define popular songs (top 25%)
popularity_threshold = df['pop'].quantile(0.75)
df['is_popular'] = df['pop'] >= popularity_threshold

# Compare features between popular and unpopular songs
features_to_compare = ['Energy', 'dance', 'val', 'acous']
popular_comparison = df.groupby('is_popular')[features_to_compare].mean()

print("\nPopular vs Unpopular Songs:")
print(popular_comparison)
```

```python
df.head()
```

```python
# Calculate percentage difference
diff = ((popular_comparison.loc[True] - popular_comparison.loc[False]) / 
        popular_comparison.loc[False] * 100)
print("\nPercentage Difference (Popular vs Unpopular):")
print(diff.sort_values(ascending=False))
```

```python
# Visualize
fig, ax = plt.subplots(figsize=(12, 6))
x = np.arange(len(features_to_compare))
width = 0.2

ax.bar(x - width/2, popular_comparison.loc[False], width, label='Unpopular', alpha=0.8)
ax.bar(x + width/2, popular_comparison.loc[True], width, label='Popular', alpha=0.8)

ax.set_xlabel('Features', fontsize=12)
ax.set_ylabel('Average Value', fontsize=12)
ax.set_title('Feature Comparison: Popular vs Unpopular Songs', fontsize=16, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(features_to_compare, rotation=45)
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# ============================================================
# 4. FEATURE RELATIONSHIPS
# ============================================================

print("\n" + "=" * 60)
print("4. EXPLORING FEATURE RELATIONSHIPS")
print("=" * 60)
```

```python
# Question 4a: Do high-energy songs tend to be more danceable?
# ------------------------------------------------------------

# Calculate correlation
energy_dance_corr = df['Energy'].corr(df['dance'])
print(f"\nCorrelation between Energy and Danceability: {energy_dance_corr:.3f}")
```

```python
# Create scatter plot
plt.figure(figsize=(12, 8))
plt.scatter(df['Energy'], df['dance'], alpha=0.3, s=20)

# Add trend line
z = np.polyfit(df['Energy'].dropna(), df['dance'].dropna(), 1)
p = np.poly1d(z)
plt.plot(df['Energy'].sort_values(), p(df['Energy'].sort_values()), 
         "r--", linewidth=2, label=f'Trend Line (r={energy_dance_corr:.2f})')

plt.xlabel('Energy', fontsize=12)
plt.ylabel('Danceability', fontsize=12)
plt.title('Relationship Between Energy and Danceability', fontsize=16, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# ============================================================
# CORRELATION MATRIX FOR ALL FEATURES
# ============================================================
```

```python
print("\n" + "=" * 60)
print("OVERALL FEATURE CORRELATION MATRIX")
print("=" * 60)
```

```python
# Select numeric columns
numeric_cols = ['Energy', 'dance', 'val', 'acous', 'spch', 'pop']

# Filter to only columns that exist in your dataset
numeric_cols = [col for col in numeric_cols if col in df.columns]

# Calculate correlation matrix
correlation_matrix = df[numeric_cols].corr()

print("\nTop 5 Strongest Correlations:")
# Get upper triangle of correlation matrix
corr_pairs = correlation_matrix.unstack()
corr_pairs = corr_pairs[corr_pairs < 1]  # Remove self-correlations
print(corr_pairs.abs().sort_values(ascending=False).head(10))
```

```python
# Visualize
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', 
            center=0, square=True, linewidths=1,
            cbar_kws={'label': 'Correlation Coefficient'})
plt.title('Feature Correlation Matrix', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
